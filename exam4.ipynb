 **1 -> Explain the purpose of version control systems like Git. Why are they
essential in modern software development?**

- The primary purposes of a VCS are to manage a projects history facilitate collaboration and ensure code integrity.
- version control systems are tools that track changes to files over time allowing teams to collaborate efficiently on software development projects. They record modifications to a projects codebase providing a history of every change.

-  They streamline the workflow for large distributed teams allowing members to work on separate features in parallel and integrate their work smoothly.

- VCS help maintain a high standard of code quality and stability.

**3 ->  What is the purpose of a .gitignore file in a Git repository? Provide an
example of how to ignore all .log files.**

- The .gitignore file helps maintain a clean and efficient repository by excluding files that are not essential to the projects source code or build process. Common files and directories typically ignored include.

- To ignore all files ending with the .log extension you would add the following line to a file named .gitignore in the root directory of your Git repository.

 *.log

- This pattern matches any file with any name as long as it ends in .log  such as app.log error.log or test results .log.
